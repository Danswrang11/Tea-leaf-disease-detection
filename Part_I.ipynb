# Btach Rename

```python
import os
import shutil

# Set the directory containing the images
directory = '/content/drive/MyDrive/Project/Dataset/Anthracnose'

# Set the starting index for the numbering
index = 1

# Loop through all the files in the directory
for filename in os.listdir(directory):
    # Check if the file is an image file
    if filename.endswith('.jpg') or filename.endswith('.jpeg') or filename.endswith('.JPG'):
        # Get the full path of the file
        filepath = os.path.join(directory, filename)

        # Generate a new filename
        new_filename = 'Anthracnose' + '_' + str(index) + '.' + filename.split('.')[1]

        # Get the full path of the new file
        new_filepath = os.path.join(directory, new_filename)

        # Rename the file
        shutil.move(filepath, new_filepath)

        # Increment the index
        index += 1
```

```python
import cv2
import os
import imquality.brisque as brisque
from brisque import BRISQUE
```

```python
# Set the paths to the folders containing the images
folder_paths = ['/content/drive/MyDrive/Project/Dataset/Algal_spot', '/content/drive/MyDrive/Project/Dataset/Anthracnose',
                '/content/drive/MyDrive/Project/Dataset/Birds eye spot','/content/drive/MyDrive/Project/Dataset/Brown blight',
                '/content/drive/MyDrive/Project/Dataset/Gray blight','/content/drive/MyDrive/Project/Dataset/Healthy',
                '/content/drive/MyDrive/Project/Dataset/Non tea leaf','/content/drive/MyDrive/Project/Dataset/Red leaf spot',
                '/content/drive/MyDrive/Project/Dataset/White spot']

# Create a BRISQUE object
brisque = BRISQUE()

# Initialize a dictionary to store the scores
scores = {}
```

```python
# Loop through each folder
for folder_path in folder_paths:
    folder_name = os.path.basename(os.path.normpath(folder_path))
    scores[folder_name] = {}

    # Loop through each image in the folder
    for filename in os.listdir(folder_path):
        if filename.endswith('.jpg') or filename.endswith('.png'): # Adjust file extension based on your images
            # Load the image
            img_path = os.path.join(folder_path, filename)
            img = cv2.imread(img_path)

            # Calculate the BRISQUE score
            score = brisque.score(img)

            # Add the score to the dictionary of scores
            scores[folder_name][filename] = score

# Print the scores
print("BRISQUE Scores:")
for folder_name, folder_scores in scores.items():
    print(f"Folder: {folder_name}")
    for filename, score in folder_scores.items():
        print(f"\t{filename}: {score:.2f}")
```

# Applying Threshold of 40

```python
# Set the paths to the folders containing the images
folder_paths = ['/content/drive/MyDrive/Project/Dataset/Algal_spot', '/content/drive/MyDrive/Project/Dataset/Anthracnose',
                '/content/drive/MyDrive/Project/Dataset/Birds eye spot','/content/drive/MyDrive/Project/Dataset/Brown blight',
                '/content/drive/MyDrive/Project/Dataset/Gray blight','/content/drive/MyDrive/Project/Dataset/Healthy',
                '/content/drive/MyDrive/Project/Dataset/Non tea leaf','/content/drive/MyDrive/Project/Dataset/Red leaf spot',
                '/content/drive/MyDrive/Project/Dataset/White spot']

# Create a BRISQUE object
brisque = BRISQUE()

# Initialize a dictionary to store the scores
scores = {}
threshold_score = 40
# Loop through each folder
for folder_path in folder_paths:
    folder_name = os.path.basename(os.path.normpath(folder_path))
    scores[folder_name] = {}

    # Loop through each image in the folder
    for filename in os.listdir(folder_path):
        if filename.endswith('.jpg') or filename.endswith('.png'): # Adjust file extension based on your images
            # Load the image
            img_path = os.path.join(folder_path, filename)
            img = cv2.imread(img_path)

            # Calculate the BRISQUE score
            score = brisque.score(img)

            # Add the score to the dictionary of scores
            scores[folder_name][filename] = score

            # Check if the score is above the threshold
            if score > threshold_score:
              # Delete the image
              os.remove(os.path.join(folder_path, filename))

# Print the scores
threshold_score1 =40
print("BRISQUE Scores:")
for folder_name, folder_scores in scores.items():
    print(f"Folder: {folder_name}")
    for filename, score in folder_scores.items():
      if score <= threshold_score1:
        print(f"\t{filename}: {score:.2f}")
```
